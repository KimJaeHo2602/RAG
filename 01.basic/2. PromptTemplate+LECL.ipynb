```python
# ì¤‘ê´„í˜¸ëŠ” ë³€ìˆ˜ì§€ë§Œ, ëŒ€ê´„í˜¸ëŠ” ë³€ìˆ˜ê°€ ì•„ë‹ˆë‹¤! ì¤‘ê´„í˜¸ë¥¼ í‘œí˜„í•˜ê³  ì‹¶ìœ¼ë©´ ë‘ë²ˆì“°ë©´ ëœë‹¤ {{}}
# #ì€ êµ¬ë¶„ìì´ë‹¤.
template = """
ë‹¹ì‹ ì€ ì˜ì–´ë¥¼ ê°€ë¥´ì¹˜ëŠ” 10ë…„ì°¨ ì˜ì–´ ì„ ìƒë‹˜ì…ë‹ˆë‹¤. ì£¼ì–´ì§„ ìƒí™©ì— ë§ëŠ” ì˜ì–´ íšŒí™”ë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”. íšŒí™”ì—ì„œ ê°ìì˜ ì—­í• ê¹Œì§€ ì•Œë ¤ì£¼ì„¸ìš”.
ì–‘ì‹ì€ [FORMAT]ì„ ì‘ì„±í•´ ì£¼ì„¸ìš”.

#ìƒí™©
{question}

#FORMAT:
- ì˜ì–´ íšŒí™”:
- í•œê¸€ í•´ì„:
"""
prompt = PromptTemplate.from_template(template)
prompt
```

```python
model = ChatOllama(model="llama3:8b")

output_parser = StrOutputParser()
```

```python
chain = prompt | model | output_parser
```
