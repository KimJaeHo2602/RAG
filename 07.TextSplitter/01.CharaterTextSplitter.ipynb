### CharacterTextSplitter

기본적으로 `\n\n`을 기준으로 문자 단위를 텍스트를 분할하고, 청크의 크기를 문자 수로 측정한다.
1. 텍스트 분할 방식: 단일 문자 기준
2. 청크 크기 측정 방식: 문자수 기준

- `./data/appendix-keywords.txt`파일을 열어 내용을 읽음
- 읽어들인 내용을 `file` 변수에 저장.

```python
with open("./data/appendix-keywords.txt", encoding='utf-8') as f:
    file = f.read()
```

```python
print(file[:500])
```

```python
print(file)
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(
    #separator="\n\n",
    chunk_size=190,
    chunk_overlap=0,
    #length_function=len
)
```

```python
texts = text_splitter.create_documents([file])
```

```python
print(texts[0].page_content)
```

```python
metadatas = [
    {"document": 1},
    {"document": 2},
]  # 문서에 대한 메타데이터 리스트를 정의합니다.
documents = text_splitter.create_documents(
    [
        file,
        file,
    ],  # 분할할 텍스트 데이터를 리스트로 전달합니다.
    metadatas=metadatas,  # 각 문서에 해당하는 메타데이터를 전달합니다.
)
print(documents[0])  # 분할된 문서 중 첫 번째 문서를 출력합니다.
```

```python
from langchain_community.document_loaders import TextLoader
docs = TextLoader("./data/appendix-keywords.txt", encoding='utf-8')
docs = docs.load()
```

```python
print(docs)
```

```python
# 지정한 190을 넘었다는 오류. 만약, chunk_size를 210을 준다면, 오류메세지를 반환하지 않을 것이다.
documents = text_splitter.split_documents(docs)
```
